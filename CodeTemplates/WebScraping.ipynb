```python
#Challenge 3
import seaborn as sns
import pandas as pd
def substringOccur(mainString, substring):
    count = 0
    start = 0
    
    while start < len(mainString):
        start = mainString.find(substring, start)
        if start == -1:
            break
        count += 1
        start += 1
        
    return count
```

```python
print(countList)
avg = sum(countList) / len(countList)
print(avg)
```
