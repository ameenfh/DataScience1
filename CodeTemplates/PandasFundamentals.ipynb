```python
#Challenge 2 part 3
# Create an empty list to store the row and column indices of missing data
missing_indices = []

# Iterate through the DataFrame to find missing data
for row_index, rowData in df.iterrows():
    for col_index, value in rowData.items():
        if pd.isna(value):
            missing_indices.append((row_index, col_index))

# Display the row and column indices of missing data
for row_Null, col_Null in missing_indices:
    print(f"Row {row_Null + 1}, Column {col_Null} has missing data.")
```

```python
#Challenge 2 part 3 (different method)

#df.columns gives out all the column names
for column in df.columns:
    for row in range(len(df)):
        #df.iloc[row index, column index] --> value of a dataframe at an index location. Indices start at 0
        #df.loc[row index, column name] --> gives value of dataframe at row index and column name.
        if pd.isna(df.loc[row, column]):
            print(row + 1, column)
```

```python
#Challenge 4 part 2
print("The average age of a pirate is " + str(df["age"].mean()) + " years old")
femaleAges = []

for i in range(len(df)):
    if df.loc[i, "sex"] == "female":
        femaleAges.append(df.loc[i, "age"])

femaleAvg = sum(femaleAges)/len(femaleAges)
        

print("The average age of a female pirate is " + str(femaleAvg) + " years old")

df["age"][df["sex"] == "female"].mean()
```

```python
#Challenge 4 part 3
count = 0

for row in range(len(df)):
    weight = df.loc[row, "weight"]
    if (weight * 2.2) < 100:
        count += 1
percentLess = (count/len(df)) * 100
print(f"The percent of pirates with weights less than 100 pounds is {percentLess}%")
```

```python
#Challenge 5 part 1
minHeight = df.at[0, "height"]
count = 0
for i in df["height"]:
    count += 1
    if i < minHeight:
        minHeight = i
        id = count
        
print(f"The shortest pirate has a height of {minHeight}cm and and id of {id}")
```

```python
#Challenge 5 part 2
maxWeight = df.at[0, "weight"]
count = 0
for i in df["weight"]:
    count += 1
    if i > maxWeight:
        maxWeight = i
        id = count
        
print(f"The shortest pirate has a height of {maxWeight}kg and and id of {id}")
```
