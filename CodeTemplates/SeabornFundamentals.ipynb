```python
#Challenge 1

import pandas as pd

df = pd.read_csv("compas.csv")

df.info()
```

```python
#Challenge 2

rowNumber = len(df)

print(f"There are {rowNumber} rows in this dataset")
print(df.loc[10, "sex"])
```

```python
#Challenge 3
import seaborn as sns
sns.countplot(x = df["sex"])
```

```python
#Challenge 4

oldCount = 0
midCount = 0
youngCount = 0



for cat in df["age_cat"]:
    if cat == "Greater than 45":
        oldCount += 1
    elif cat == "25 - 45":
        midCount += 1
    else:
        youngCount += 1

print(oldCount)
print(midCount)
print(youngCount)

sns.countplot(x = df["age_cat"])
```

```python
#Challenge 6
sns.histplot(x = df["age"], kde = True, color = "green", bins = 8)

#kde is the  kernel density estimation curve
#bins is the number of bins
```

```python
#Challenge 7
import matplotlib.pyplot as plt

df["c_jail_in"] = pd.to_datetime(df["c_jail_in"])
df["c_jail_out"] = pd.to_datetime(df["c_jail_out"])

#new columnn
df["total_jailtime"] = (df["c_jail_out"] - df["c_jail_in"]).dt.days


sns.set_style("whitegrid")
sns.scatterplot(x = df["total_jailtime"], y = df["age"])

import matplotlib.pyplot as plt
plt.xlabel("Days in Jail")
plt.ylabel("Age")
plt.title("Days in Jail after First Charge")
plt.show()
```
