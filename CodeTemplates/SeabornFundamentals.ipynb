```python
#Challenge 3
import seaborn as sns
sns.countplot(x = df["sex"])
```

```python
#Challenge 4

oldCount = 0
midCount = 0
youngCount = 0



for cat in df["age_cat"]:
    if cat == "Greater than 45":
        oldCount += 1
    elif cat == "25 - 45":
        midCount += 1
    else:
        youngCount += 1

print(oldCount)
print(midCount)
print(youngCount)

sns.countplot(x = df["age_cat"])
```
